Next, Let us consider more general model, Ferromagnetic model with Dzyaloshinsky-Moriya interaction and single ion anisotropy.\\
The model can rewrite as follow:\\
\begin{equation}
H = H_{0} + H_{DMI} + H_{SIA}\\
\\
H_{0} = J_{ij}\sum_{<i,j>}{S_{i}\cdot S_{i}}\\
H_{DMI} =  \sum_{<i,j>}D_{ij}\cdot(S_{i} \times S_{j})\\
H_{SIA} = -\frac{1}{2} \sum_{<i>}{\delta (S_{i}^{z})^{2}}
\end{equation}

We can rewrite it as second quantum form by HP transformation:
\begin{equation}
H_{0} = JS\sum_{i,j}{(\hat{a}_{i}^{+}-\hat{b}_{j}^{+})(\hat{a}_{i}-\hat{b}_{j})}\\
H_{DMI} = \frac{-iDS}{2}(\sum_{<i,\delta>}\hat{e}(\delta){(-\hat{a}_{i}\hat{a}^{+}_{i+\delta} + \hat{a}^{+}_{i}\hat{a}_{i+\delta})} + \sum_{<j,{\delta}^{'}>}\hat{e}({\delta}^{'}){(-\hat{b}_{j}^{+}\hat{b}_{j+\delta^{'}} + \hat{b}^{+}_{j}\hat{b}_{j+\delta^{'}} )})\\
H_{SIA} = S\delta (\sum_{<i>}\hat{a}_{i}^{+}\hat{a}_{i} + \sum_{<j>}\hat{b}_{j}^{+}\hat{b}_{j})
\end{equation}

For a lattice with periodic boundary conditions, $a_{i+N} = a_{i}$, we introduce the canonical bosonic operators, $\hat{a}_{k}$ and $\hat{a}^{+}_{k}$, which create and annihilate magnons with momentum k, respectively:

\begin{equation}
\hat{a}^{+}_{k} = \sqrt{\frac{1}{N}}\sum_{k}e^{-ik\cdot{r_{i}}}\hat{a}_{i}^{+}\\
\hat{a}_{k} = \sqrt{\frac{1}{N}}\sum_{k}e^{ik\cdot{r_{i}}}\hat{a}_{i}
\end{equation}

Now, we can rewrite the above Hamiltonian ainto k space:
\begin{equation}
H_{0} = -JSz \sum_{k}(\gamma_{k}\hat{a}_{k}\hat{b}_{k}^{+} + \gamma_{k}^{*}\hat{a}_{k}^{+}\hat{b}_{k} - \hat{a}_{k}^{+}\hat{a}_{k} - \hat{b}_{k}^{+}\hat{b}_{k})\\
H_{DMI} = DS\sum_{k}f_{k}(\hat{a}_{k}^{+}\hat{a}_{k} - \hat{b}_{k}^{+}\hat{b}_{k})\\
H_{SIA} = \delta S\sum_{k}(\hat{a}_{k}^{+}\hat{a}_{k} + hat{b}_{k}^{+}\hat{b}_{k})\\
\end{equation}

In the above equation, $z=(3)$ is the number of NN of each lattice site. $\gamma_{k} = \frac{1}{z} \sum_{\sigma}e^{-ik\cdot{\sigma}}$, and $f_{k} = 2\sum_{\mu}sin(k \cdot \mu)$\\                                         


Then, we can rewrite the total Hamiltonian in a matrix form:
\begin{equation}
H_{k} = JS\sum_{k} \Psi(k)^{\dagger}M_{k} \Psi(k)
\end{equation}
with $\Psi(k) = (\hat{a}_{k}, \hat{b}_{k})$ a two-component wavefunction and 
\begin{equation}
M_{k} = \left(
  \begin{array}{ccc}   
    {3+\delta^{'} + D^{'}f_{k}} & -\gamma_{k} \\ 
    -\gamma_{k}^{*} & {3+\delta^{'} - D^{'}f_{k}} \\ 
  \end{array}
\right)    
\end{equation}

where $\delta^{'} = \delta / J$ and $D^{'} = D/J$. 

In terms of the pauli matrix form, $M(k)$ can be rewritten as:
\begin{equation}
M(k) = h_{0}(k)\sigma_{0} + h_{x}(k)\sigma_{x} + h_{y}(k)\sigma_{y} + h_{z}(k)\sigma_{z}
\end{equation}

where $h_{0}(k) = 3 + \delta^{'}, h_{x}(k) = -\sum_{\sigma} {cos(k\cdot \sigma)}, h_{y}(k) = -\sum_{\sigma} {sin(k\cdot \sigma)}, h_{z}(k) = D^{'}f_{k}$\\

Therefore, the corresponding eigenvalues is:
\begin{equation}
\epsilon(k) = h_{0}(k) \pm \sqrt{(h_{x}(k)^{2} + h_{y}(k)^{2} + h_{z}(k)^{2})}\\
\epsilon(k) = h_{0}(k) \pm \sqrt{(h_{x}(k)^{2} + h_{y}(k)^{2} + h_{z}(k)^{2})}
\end{equation}
